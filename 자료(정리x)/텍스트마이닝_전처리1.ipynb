```python
import pandas as pd

rest = pd.read_csv('/content/drive/MyDrive/서울시 일반음식점 인허가 정보.csv', encoding='EUC-KR', encoding_errors='ignore', low_memory=False)
rest = rest[['상세영업상태명','도로명주소','도로명우편번호','사업장명','업태구분명','좌표정보(X)','좌표정보(Y)']]
rest = rest[rest['상세영업상태명']=='영업']
rest = rest.drop(columns='상세영업상태명')
```

```python
value_to_remove = ['출장조리','키즈카페','이동조리','감성주점','라이브카페']

if '업태구분명' in rest.columns:
    rest = rest[~rest['업태구분명'].isin(value_to_remove)]
```

```python
# 특정 열(column)을 기준으로 위아래로 내용의 절반 이상이 동일한 행 제거하고 하나만 남기기
column_to_check = '도로명주소'

if column_to_check in rest.columns:
    # 중복 행 제거
    rest = rest.drop_duplicates(subset=column_to_check)

    # 하나만 남기기
    rest = rest.groupby(column_to_check).first().reset_index()
```

```python
seoul_districts = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구',
                   '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구',
                   '용산구', '은평구', '종로구', '중구', '중랑구']

def extract_district(address):
    for district in seoul_districts:
        if district in address:
            return district
    return ''

# 가정집 주소를 기반으로 지역구 추출하여 지역구 column 추가
rest['지역구'] = rest['도로명주소'].apply(extract_district)

rest
```

```python
cul1 = pd.read_csv('/content/drive/MyDrive/서울시 관광 문화.csv', encoding='EUC-KR', encoding_errors='ignore', low_memory=False)
cul2 = pd.read_csv('/content/drive/MyDrive/서울시 문화공간 정보.csv', encoding='EUC-KR', encoding_errors='ignore', low_memory=False)
cul1 = cul1[['상호명','신주소','웹사이트']]
cul2 = cul2[['문화시설명','주소','홈페이지','주제분류','위도','경도']]
cul1.columns = ['문화시설명','주소','홈페이지']
```

```python
cul1['주소'] = cul1['주소'].astype(str)
cul2['주소'] = cul2['주소'].astype(str)

seoul_districts = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구',
                   '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구',
                   '용산구', '은평구', '종로구', '중구', '중랑구']

def extract_district(address):
    for district in seoul_districts:
        if district in address:
            return district
    return ''

# 가정집 주소를 기반으로 지역구 추출하여 지역구 column 추가
cul1['지역구'] = cul1['주소'].apply(extract_district)
cul2['지역구'] = cul2['주소'].apply(extract_district)

cul1 = cul1[cul1['지역구']!='']
cul2 = cul2[cul2['지역구']!='']
```

```python
display(cul1,cul2)
```

```python
def find_common_elements(list1, list2):
    return list(set(list1) & set(list2))

common_elements1 = find_common_elements(cul1['홈페이지'].unique(), cul2['홈페이지'].unique())
common_elements1 = common_elements1[1:]
common_elements2 = find_common_elements(cul1['문화시설명'].unique(), cul2['문화시설명'].unique())

if '홈페이지' in cul1.columns:
    cul1 = cul1[~cul1['홈페이지'].isin(common_elements1)]

if '문화시설명' in cul1.columns:
    cul1 = cul1[~cul1['문화시설명'].isin(common_elements2)]
```

```python
cul = pd.concat([cul1,cul2],axis=0)
```

```python
gu_list = ['송파구','중구','종로구','용산구','강남구','서초구','서대문구','노원구','동대문구','마포구','영등포구','성동구','용산구','광진구']
if '지역구' in cul.columns:
  cul = cul[cul['지역구'].isin(gu_list)]
```

```python
cul['검색항목'] = cul['문화시설명'] +' '+ cul['지역구']
cul
```

```python
cul.to_csv('장소 원본.csv',index=0)
```
